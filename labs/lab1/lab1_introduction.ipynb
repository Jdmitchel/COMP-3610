````python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
````

````python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(df['Year'], df['burglary'], label='burglary', color='blue')
ax.plot(df['Year'], df['assault'], 'o', label='assault', color='orange')
ax.annotate('maximum', xy=(1978.0, 1878), xytext=(1980.0, 1878), arrowprops=dict(facecolor='red', shrink=0.05))
ax.set_xlabel('Year')
ax.set_ylabel('Number of crimes')
ax.legend()
````

- Looking at the dataset we can see that Survived and Pclass are both categorical but in numeric form already so we don't change these.

````python
newClass = pd.Categorical(df_titanic["Pclass"], ordered=True)
newClass = newClass.rename_categories(["class1", "class2","class3"])
newClass.describe()
````

Let's look at the Cabin column

````python
df_titanic["Cabin"].unique()
````

- Each section of thee ship is represented by a capital letter in the Cabin name, we can group by these letters in order to make this column more useful.

````python
cabinStr = df_titanic["Cabin"].astype(str) #converts the data to str
newCabin = np.array([cabin[0] for cabin in cabinStr]) #Extract first letter
newCabin = pd.Categorical(newCabin)
df_titanic["Cabin"] = newCabin
df_titanic["Cabin"].describe()
````

````python
newCabin.describe()
````

````python
df_titanic
````

- Cabin has alot of missing data but cabin location may impact survival rate so we can keep it as is for now.
- Age has missing data but instead of just getting rid of those rows (177 rows), let's instead keep them and replace the missing data with the mean value, but first let's look at the data for Age.
- We also introduce the **seaborn** package. Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

### Useful Resource:
